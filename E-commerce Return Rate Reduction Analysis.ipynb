# EDA

```python
#TOTAL ORDERS
total_orders = df['Order_ID'].nunique()

print("Total Orders:", total_orders)
```

```python
#TOTAL REVENUE
total_revenue = df['Order_Value'].sum()

print("Total Revenue:", round(total_revenue,2))
```

```python
#TOTAL PROFIT / LOSS
total_profit = df['Profit_Loss'].sum()

print("Total Profit/Loss:", round(total_profit,2))
```

```python
#AVERAGE ORDER VALUE
avg_order = df['Order_Value'].mean()

print("Average Order Value:", round(avg_order,2))
```

```python
#PRODUCT CATEGORY ANALYSIS
category = df.groupby('Product_Category')['Order_Value'].sum().sort_values(ascending=False)

category
```

```python
category.plot(kind='bar', figsize=(10,5))

plt.title("Revenue by Product Category")
plt.xlabel("Category")
plt.ylabel("Revenue")

plt.show()
```

Shows the total revenue generated by each product category.

```python
#SHIPPING METHOS DISTRIBUTION
shipping = df['Shipping_Method'].value_counts()

shipping
```

```python
shipping.plot(kind='bar')

plt.title("Shipping Method Distribution")

plt.show()
```

Shows the number of orders for each shipping method.

```python
#PAYMENT METHOD ANALYSIS
payment = df['Payment_Method'].value_counts()

payment
```

```python
payment.plot(kind='pie',
             autopct='%1.1f%%')

plt.ylabel("")

plt.title("Payment Method Distribution")

plt.show()
```

Shows the percentage distribution of different payment methods used by customers.

```python
#GENDER DISTRIBUTION
gender = df['User_Gender'].value_counts()

gender
```

```python
gender.plot(kind='bar')

plt.title("Gender Distribution")

plt.show()
```

Shows the distribution of customers based on gender.

```python
#AGE DISTRIBUTION
plt.figure(figsize=(8,5))

plt.hist(df['User_Age'], bins=20)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

Shows the distribution of customers across different age groups.

```python
#TOP 10 USER LOCATIONS
location = df['User_Location'].value_counts().head(10)

location
```

```python
location.plot(kind='bar')

plt.title("Top 10 User Locations")

plt.show()
```

Shows the top 10 locations with the highest number of customers or orders.

```python
#MONTHLY REVENUE TREND
monthly = df.groupby(df['Order_Date'].dt.to_period('M'))['Order_Value'].sum()

monthly
```

```python
monthly.plot(figsize=(12,5), marker='o')

plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")

plt.show()
```

Shows how total revenue changes over different months.

```python
#RETURN STATUS
returns = df['Return_Status'].value_counts()

returns
```

```python
returns.plot(kind='bar')

plt.title("Return Status")

plt.show()
```

Shows the number of returned and non-returned orders.

```python
#RETURN REASONS
reason = df['Return_Reason'].value_counts()

reason
```

```python
reason.plot(kind='bar', figsize=(10,5))

plt.title("Return Reasons")

plt.show()
```

Shows the most common reasons why customers return products.

```python
#AVERAGE DAYS TO RETURN
avg_days = df['Days_to_Return'].mean()

print("Average Days to Return:", round(avg_days,2))
```

```python
#PROFIT BY CATEGORY
profit = df.groupby('Product_Category')['Profit_Loss'].sum()

profit
```

```python
profit.plot(kind='bar')

plt.title("Profit by Category")

plt.show()
```

Shows the total profit or loss generated by each product category.

```python
#CORRELATION ANALYSIS
corr = df[['Product_Price',
           'Order_Quantity',
           'Order_Value',
           'Profit_Loss',
           'Return_Cost',
           'CO2_Emissions',
           'Packaging_Waste',
           'CO2_Saved',
           'Waste_Avoided']].corr()

corr
```

```python
import seaborn as sns

plt.figure(figsize=(10,7))

sns.heatmap(corr,
            annot=True,
            cmap='coolwarm')

plt.title("Correlation Matrix")

plt.show()
```

Shows the relationship and correlation between numerical variables in the dataset.

```python
#PRODUCT PRICE DISTRIBUTION
plt.figure(figsize=(8,5))

plt.hist(df['Product_Price'], bins=30)

plt.title("Product Price Distribution")

plt.show()
```

Shows how product prices are distributed across all products.

```python
#ORDER QUANTITY DISTRIBUTION
plt.figure(figsize=(8,5))

plt.hist(df['Order_Quantity'], bins=20)

plt.title("Order Quantity Distribution")

plt.show()
```

Shows the frequency of different order quantities placed by customers.

```python
#ORDER VALUE VS PROFIT
plt.figure(figsize=(8,5))

plt.scatter(df['Order_Value'],
            df['Profit_Loss'])

plt.title("Order Value vs Profit")

plt.xlabel("Order Value")

plt.ylabel("Profit/Loss")

plt.show()
```

Shows the relationship between order value and profit or loss.

```python
#SUSTAINABILITY ANALYSIS
sustainability = df[['CO2_Emissions',
                     'CO2_Saved',
                     'Packaging_Waste',
                     'Waste_Avoided']].sum()

sustainability
```

```python
sustainability.plot(kind='bar')

plt.title("Environmental Impact Summary")

plt.ylabel("Total")

plt.show()
```

Shows the overall environmental impact through CO₂ emissions, CO₂ savings, packaging waste, and waste avoided.
